```python
import re
import matplotlib.pyplot as plt
from collections import defaultdict

def parse_log_file(filepath):
    epoch_losses = defaultdict(list)
    ndcg_scores = {}
    hit_scores = {}

    with open(filepath, 'r') as f:
        for line in f:
            # Parse loss lines
            loss_match = re.match(r'loss in epoch (\d+) iteration \d+: ([\d\.]+)', line)
            if loss_match:
                epoch = int(loss_match.group(1))
                loss = float(loss_match.group(2))
                epoch_losses[epoch].append(loss)

            # Parse ndcg@10
            ndcg_match = re.search(r'epoch (\d+).*ndcg@10[:=]?\s*([\d\.]+)', line, re.IGNORECASE)
            if ndcg_match:
                epoch = int(ndcg_match.group(1))
                ndcg_scores[epoch] = float(ndcg_match.group(2))

            # Parse hit@10
            hit_match = re.search(r'epoch (\d+).*hit@10[:=]?\s*([\d\.]+)', line, re.IGNORECASE)
            if hit_match:
                epoch = int(hit_match.group(1))
                hit_scores[epoch] = float(hit_match.group(2))

    # Average loss per epoch
    avg_loss = {epoch: sum(losses) / len(losses) for epoch, losses in epoch_losses.items()}
    return avg_loss, ndcg_scores, hit_scores

def plot_metric(metric1, metric2, ylabel, title, label1='Beauty', label2='ML-1M'):
    plt.figure(figsize=(8, 5))
    x1, y1 = zip(*sorted(metric1.items()))
    x2, y2 = zip(*sorted(metric2.items()))
    plt.plot(x1, y1, marker='o', label=label1)
    plt.plot(x2, y2, marker='s', label=label2)
    plt.xlabel('Epoch')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Replace with actual paths if files are in different location
beauty_file = "output_beauty_200.log"
ml_file = "output_ml-m1_200.log"

beauty_loss, beauty_ndcg, beauty_hit = parse_log_file(beauty_file)
ml_loss, ml_ndcg, ml_hit = parse_log_file(ml_file)
```

```python
plot_metric(beauty_loss, ml_loss, ylabel="Loss", title="Epoch vs. Loss")
# plot_metric(beauty_ndcg, ml_ndcg, ylabel="NDCG@10", title="Epoch vs. NDCG@10")
# plot_metric(beauty_hit, ml_hit, ylabel="Hit@10", title="Epoch vs. Hit@10")


if beauty_ndcg and ml_ndcg:
    plot_metric(beauty_ndcg, ml_ndcg, ylabel="NDCG@10", title="Epoch vs. NDCG@10")
else:
    print("NDCG@10 data not found in either log file.")

if beauty_hit and ml_hit:
    plot_metric(beauty_hit, ml_hit, ylabel="Hit@10", title="Epoch vs. Hit@10")
else:
    print("Hit@10 data not found in either log file.")
```
